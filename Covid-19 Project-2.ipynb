```python
#here we have nation wide daily covid-19 cases. We analyse this data by using different libraries.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option("display.max_rows",None,"display.max_columns",None)
```

```python
daily_cases=pd.read_csv("nation_level_daily.csv")
```

```python
daily_cases
```

```python
mask1=daily_cases["Date"].str.contains("January")
jan=daily_cases[mask1]
```

```python
jan
```

```python
#total cases in january
```

```python
jan["Total Confirmed"].sum()
```

```python
plt.style.use("classic")
```

```python
plt.rcParams["figure.figsize"]=20,10
```

```python
jan.plot(x="Date",y="Total Confirmed")
```

```python
mask2=daily_cases["Date"].str.contains("February")
feb=daily_cases[mask2]
```

```python
feb
```

```python
#total cases in february
```

```python
feb["Total Confirmed"].sum()
```

```python
feb[["Date","Total Confirmed"]].sort_values("Total Confirmed",ascending=False).style.background_gradient(cmap="Reds")
```

```python
feb.plot(kind="barh",x="Date",y="Total Confirmed")
```
