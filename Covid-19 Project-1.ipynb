```python
#performance of states
```

```python
def classify_performance(number):
    if number<50000:
        return "Good Performance"
    elif number>=50000 and number<100000:
        return "Average Performance"
    else:
        return "Bad Performance"
```

```python
performance=new_state_wise_cases["Confirmed"].apply(classify_performance)
```

```python
performance.to_frame()
```

```python
available_states.merge(performance,how="inner",left_index=True,right_index=True)
```
