```python
#here we have Covid-19 tests day wise data. We analyse this data by using different libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option("display.max_rows",None,"display.max_columns",None)
```

```python
test_day_wise=pd.read_csv("tests_day_wise.csv")
```

```python
test_day_wise
```

```python
type(test_day_wise)
```

```python
#the dataframe has many irrelevant columns. So we drop those irrelevant columns.
```

```python
new_test_day_wise=test_day_wise.drop(["Tested As Of","Total Individuals Tested","Tests conducted by Private Labs","Sample Reported today","Positive cases from samples reported","Source","Source 1","Individuals Tested Per Confirmed Case","Tests Per Confirmed Case"],axis=1)
```

```python
new_test_day_wise
```

```python
#as the dataset have empty cells we drop it by using dropna method.
```

```python
new_test_day_wise.dropna(how="any",inplace=True)
```

```python
new_test_day_wise
```

```python
#highest number of total samples tested dates
```

```python
new_test_day_wise[["Update Time Stamp","Total Samples Tested"]].sort_values("Total Samples Tested",ascending=False).style.background_gradient(cmap="Greens")
```

```python
#graphical representation
```

```python
plt.rcParams["figure.figsize"]=20,8
```

```python
plt.style.use("classic")
```

```python
new_test_day_wise.plot(kind="bar",y="Total Samples Tested")
```

```python
new_test_day_wise.head(3)
```

```python
#highest test positivity rate dates
```

```python
highest_positivity_rate=new_test_day_wise[["Update Time Stamp","Test positivity rate"]].sort_values("Test positivity rate",ascending=False)
```

```python
highest_positivity_rate
```

```python
new_test_day_wise.head(3)
```

```python
#highest tests per million dates
```

```python
new_test_day_wise[["Update Time Stamp","Tests per million"]].sort_values("Tests per million",ascending=False).style.background_gradient(cmap="Greens")
```
